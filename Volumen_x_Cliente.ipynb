```python
import pandas as pd
import numpy as np
```

```python
UO = 'MR0301'
unidad = 'Ingreso Neto'

vol_2025 = pd.read_csv('{}_ene25_abr25.csv'.format(UO))
vol_2024 = pd.read_csv('{}_may24_dic24.csv'.format(UO))

vol_full = pd.concat([vol_2025,vol_2024],ignore_index = True)

vol_full.info()
```

```python
vol_full.drop(columns = ['[Clientes].[Cve Cliente].[Cve Cliente].[MEMBER_UNIQUE_NAME]',
                         '[Tiempo].[Cve Mes].[Cve Mes].[MEMBER_UNIQUE_NAME]',
                         '[Tiempo].[Año natural].[Año natural].[MEMBER_UNIQUE_NAME]'], inplace = True)

vol_full.info()
```

```python
vol_full.columns = ['Cliente','Mes','Año','Ingreso Neto','CFC','CF','CU']
vol_full['Mes'] = vol_full['Mes'].apply(lambda x: f"{int(x):02}")
vol_full['Año_Mes'] = vol_full['Año'].astype(str) + "-" + vol_full['Mes'].astype(str)

vol_full.info()

vol_full.head()
```

```python
vol_td = vol_full.pivot_table(index = 'Cliente', columns = 'Año_Mes', values = unidad, aggfunc = 'sum', fill_value = '')

print(vol_td.head())
```

```python
vol_final = vol_td.reset_index()
vol_final.head()
```

```python
vol_final.to_csv(f'Año_Movil_{UO}_may24_abr25_{unidad}.csv', index=False)
```
